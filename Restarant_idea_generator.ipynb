### Step 1: Store your API key in Colab Secrets

1.  **Open the Secrets Manager**: Click the "ðŸ”‘" (key) icon in the left sidebar of your Colab notebook.
2.  **Add a new secret**: Click "Add new secret".
3.  **Name the secret**: In the "Name" field, type `OPENAI_API_KEY` (or any other name you prefer, but remember it for the next step).
4.  **Paste your API key**: In the "Value" field, paste your actual OpenAI API key.
5.  **Save**: Close the secrets manager. Make sure the "Notebook access" toggle is enabled for your notebook.

### Step 2: Access the API key in your Colab notebook

Use the `google.colab.userdata` module to retrieve your stored secret. This will prevent your API key from being hardcoded in your notebook.


```python
from google.colab import userdata
import os

# Retrieve the API key from Colab Secrets
openai_api_key = userdata.get('OPENAI_API_KEY')

# Set it as an environment variable for the OpenAI library
os.environ["OPENAI_API_KEY"] = openai_api_key

print("OpenAI API key loaded and set as environment variable.")
```
