```python
from google.colab import userdata
from langchain_groq import ChatGroq
import os

# Retrieve the API key from Colab Secrets
openai_api_key = userdata.get('GROQ_API_KEY')

# Setup API Key from Colab Secrets
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')

# 3. Initialize the Free Model (Llama 3)

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.6
)

response = llm.invoke("I want to open a restaurant for Indian food. Suggest a fancy name for this.")

print(response)
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Modern way: Use .from_template to automatically find {cuisine}
prompt_template_name = PromptTemplate.from_template(
    "I want to open a restaurant for {cuisine} food. Suggest a fancy name for this. Give me the perfect one name only! And don't include any description to it"
)
name_chain = prompt_template_name | llm | StrOutputParser()


prompt_template_food_items = PromptTemplate.from_template(
    "Suggest some menu items for {restaurant_name}. Return it as a comma seperated"
)
food_items_chain = prompt_template_food_items | llm | StrOutputParser()
```

```python
from langchain_core.runnables import RunnablePassthrough

sequential_chain = (
    {"cuisine" : RunnablePassthrough()} | RunnablePassthrough.assign(restaurant_name=name_chain) | RunnablePassthrough.assign(food_items=food_items_chain)
)

result = sequential_chain.invoke("Germany")

print(result)
```
